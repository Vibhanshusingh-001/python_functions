```python
# To check number is odd and even 
num=35
def evenodd(num):
    if num%2==0:
        print("this number is even")
    else:
        print("this number is odd") 
    return        
```

```python
### Variable Length Arguments
## Positional And Keywords arguments
def printnumbers(*vb):
    for number in vb:
        print(number)
```

```python
printnumbers(2,3,4,5,6,7,8,9,"vibhu")
```

```python
# positional argument 
def numbers(*args):
    for number in args:
        print(number)
```

```python
numbers(2,3,4,5,6,7,8,9,0,"vibhu")
```

```python
# keyword arguments
def atul(**args):
    for key,value in args.items():
        print(f"{key} :{value} ")
```

```python
atul(name="vb",sub="bioinfo")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"vb",name="Krish",age="23",country="India")
```
